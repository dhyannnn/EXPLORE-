Brandt then showed several plots with several values of total dark matter density, and what happens to the cluster's radius. Essentially every value of dark matter density displayed the same pattern, the cluster starts off small and expands slowly until its density matches the dark matter density and then its radius starts to expand like $ r_h \approx \sqrt{t} $ where t is time in Gyr
