## <font color=red> General Diffusion Coefficient Derivation from Textbook </font>
<br>
- As mentioned previously the textbook gives a general derivation of the diffusion coefficient using the example of a subject star moving through a sea of field stars. Assuming the subject stars and field stars have velocities $\textbf{v}$ and $\textbf{v} _a$ respectively The relative velocity is given by $V = \textbf{v} - \textbf{v} _a $, the initial relative velocity is $V_0$ and the change in velocity due to an interaction between the subject star and the field of stars is given by $ \Delta V$. The change in velocity of the subject star $ \Delta \textbf{v}$ in terms of the change in relative change in velocity $ \Delta V$ is given by the equation <br><br> $$ \Delta \textbf{v} = \frac{m_a}{m + m_a} \Delta V ,$$ <br> where $m$ and $m_a$ are the mass of the subject star and mass of the field stars respectively. 
<br>
- The textbook then introduces a coordinate system $\hat{\textbf{e'}_1}$ , $\hat{\textbf{e'}_2}$ , $\hat{\textbf{e'}_3}$ such that the direction of the relative velocity $ \Delta V_0 $ is parallel to $\hat{\textbf{e'}_1}$ from that point onwards the derivation begins to get complicated, and I will have to read it over a couple more times to thouroughly understand it.
