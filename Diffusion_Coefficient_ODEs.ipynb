```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp 
```

```python
# Defining the ODEs 
def f(t,y): 
  vel,width = y 
  return np.array(0.01*t*width, 2*np.sqrt(width*vel))
t = np.linspace(0,10)
sol = solve_ivp(f, [0,10], [5,10], t_eval = t)

velocity = sol.y[0]
width = sol.y[1]
t = sol.t

plt.plot(t, velocity, label = 'velocity dispersion')
plt.plot(t, width, label = 'width dispersion')
plt.legend()
plt.show()
```
