```python
#Tutorial 2
#Control Structures 
#As the name suggests, control structures are the building blocks of programming. They are used to control the flow of the program. There are three types of control structures.
#1. for-loops 
#2. while-loops
#3. conditional statements 
#In the following we will expand on all of them starting with conditional statements
```

```python
#Conditional Statements 
#A conditional statement is a piece of code that is executed in a program if a certain condition is met, otherwise that part of the code is skipped over and not executed. 
#The simple statement is the if statement. It has the following basic structure in python:
# if <condition>:
#    <statement>
#In this structure:
#condition is an expression that evaluates to a Boolean value.
#statement is a valid python statement, which must be indented.
#If condition is true, then statement is executed If condition is false, then statement is skipped over, and not executed. Note the colon (:) following the condition is required.  
#Example: 
x = 5
if x>3: 
    print("x is greater than 3")
```

```python
#We can chain multiple conditions together as follows: 
x = 4
if x >3: 
    print("x is greater than 3")
    if x % 2 == 0:
        print("x is even")
    if x % 2 == 1:
        print("x is odd")
```

```python
#Now that we have learnt how to execute a code if a condition is met, we ask what else can be done? Sometimes, we want to evaluate a condition and take one path if it's true but specify an alternative path if it's not. 
#else is precisley another conditional statement.
#In python this is accomplished with an if and an else clause as follows: 
# if <condition>:
#    <statement>
# else:
#    <statement>
#In this code, if condition is true, the first suite is executed and the second is skipped. If the condition is false the first suite is skipped and the second is executed. 
#Either way execution resumes after the second suite.
#Example
x = 3
if x % 3 == 0:
    print("x is divisible by 3")
else: 
    print("x is not divisible by 3")
```

```python
#In the case we have several alternatives in the code, we use one or more elif clauses.
#Python evaluates each condition in turn and executes the suite corresponding to the first that is true. If none of the expressions are true, and an else clause is specified, then its suite is executed.
#In python this is implemented in the following way: 
# if <condition>:
#    <statement(s)>
# elif <condition>:
#    <statement(s)>
# elif <condition>:
#    <statement(s)>
#      ...
# else: 
#    <statement(s)>
#Example:
name = 'Dhyan' #change this imput and see what happens
if name == 'Fred' or name == 'fred':
     print('Hello Fred')
elif name == 'Xander' or name == 'xander':
    print('Hello Xander')
elif name == 'Joe' or name == 'joe':
     print('Hello Joe')
elif name == 'Arnold' or  name == 'arnold':
    print('Hello Arnold')
elif name == 'Dhyan' or name == 'dhyan':
    print('What up D')
else:
    print("I don't know who you are!")
```

```python
#To end this section on conditional statements, let us mention that conditional statements can be mentioned in one line, this is analog to List Comprehensions as we will see below.
#Let us see an example of how this works.
import numpy as np
x = -5
x = np.sqrt(x) if x >= 0 else np.sqrt(-x)
print(x)
#The one line if statement can be easiest understood if we read it from left to right, it can be a little tricky to get used to the syntax but if used at the right places it can make the program code much better.
```

```python
#Exercise 1 
#Conditional statements can be used anywhere in programming, particular inside functions.
#In this exercise, you are asked to define a piecewise function and plot it. The function will take a constant value of of -e if the indepndent variable is smaller than 1, it takes a value of pi if the independent variable is equal to 1 and it takes a value of sin(x) if x > 1.
#Do a scattered plot for the function in the interval -5 < x < 5, for an x-grid with logarithmic spacing. 
import numpy as np
def y(x):
    if x < 1:
        return -1*np.e
    elif x == 1:
        return np.pi
    else:
        return np.sin(x)
x = np.arange(-10,10)
y = list(map(y,x))
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#For loops 
#In this part of the tutorial we will now learn how to perform definite iteration with a python for loop 
#An iteration is a repetitive exectution of the same block of code over and over. There are two types of iteration: 
#1. Definite iteration, in which the number of repitions is specified explicitly in advance 
#2. Indefinite iteration, in which the code block executes until some condition is met in Python.
#Definite iterations are acheived using for loops, where as indefinite iterations are achieved using while loops. 
#A python for loop has the following structure:
#for <var> in <iterable>:
#    <statement(s)>
#The loop variable var is an iterator which takes on the value of the next element in iteratable. The latter corresponds to a collection of objects, for example a list, or a tuple, an interval and array, etc.
#The statement(s) in the loop body are denoted by indentation and are executed once fro each item in iteratable. 
#Example: 
for i in range(10):
    print(i+1)
```

```python
#As we learnt from Exercise 1, conditional statements can be use anywhere in a code, and in particular, they are useful in loops. Let us illustrate this with the following example:
#We are given the array of integer numbers from 0 to 100, and we want to separate the even and the odd numbers into two different list: We can do that with a for-loop as follows
given_array = np.arange(101)
even  = []
odd = []

for k in given_array:
    if k % 2 == 1:
        odd.append(k)
    else:
        even.append(k)
        
print("The list of even numbers is:", even)
print()
print("The list of odd numbers is:", odd)
```

```python
prime = []
unprime = []

for num in range (2,101):
    isPrime = True
    for i in range(2,num):
        if (num%i ==0):
            unprime.append(num)
            isPrime = False
            break
        if isPrime:
            prime.append(num)

print ("Primes")
for primeNum in prime:
    print (primeNum)
    
print ("Non-Primes")
for nonPrimeNum in unprime:
    print (nonPrimeNum)
```

```python
#While loops
#So far we have learnt how to perform definite iterations using for-loops, for this part of the tutorial we will now learn how to perform indefinite iterations using while loops. 
#A while loop has the following structure: 
#while <condition>:
#      <statement(s)>
#Likewise for for loops, statement(s) represents the block to be repeatedly executed. 
#The controlling condition, typically involves one or more variables that are initialized prior to starting the loop and then modified somewhere in the loop body.
#When a while loop is encountered, condition is first evaluated in Boolean context. If it is true the loop body is executed, then condition is checked again, and if still true the body is executed again.
#This continues until conditionbecomes false, at which point program execution proceeds to the first statement beyond the loop body.
#Example: 
x = 10
while (x > 3):
    print(x)
    x -= 1

#As you can see the condition is checked before the loop body is executed. It starts with the keyword while. After the keyword you write the condition (ie, x > 3) 
#It can be expression that evaluates to a boolean value, followed by a colon. The code is executed if the condition is true print("x is greater than 3")
```

```python
#Example:
x = 10
bigger_3 = True
while bigger_3:
    print(x)
    x = x - 1
    if x < 3:
        break 
```

```python
#Let us finally mention that every for loop can be written as a while loop: Let us see this in one of our examples above which separates even and odd numbers in the range of 0-100 
x = 100
even  = []
odd = []

while x >= 0:
    if x % 2 == 1:
        odd.append(x)
    else:
        even.append(x)
    x -= 1
even.sort() # sort() function just orders the numbers from the smallest to the biggest
odd.sort()
print("The list of even numbers is:", even)
print()
print("The list of odd numbers is:", odd)
```

```python
x = 100
i = 2
prime = []
unprime = []
while x >= 1:
    x = x - 1
    isPrime = True
    while i <= x:
        i = i + 1
        if i % x == 0:
            unprime.append(i)
            isPrime = False
    if isPrime:
        prime.append(i)
print("All the prime numbers are: ", prime)
print()
print("All the unprime numbers are:", unprime)
```
